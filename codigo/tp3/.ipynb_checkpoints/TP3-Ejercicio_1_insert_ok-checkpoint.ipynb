# Definicion de Tablas
![Vineria](diagramas/Vineria.png)


```python
Base = declarative_base()
class Customer(Base):
    __tablename__ = 'Customer'    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    address = Column(String)
    phone = Column(String)
    bday = Column(Date)
    gender = Column(String)

class Timestamp(Base):
    __tablename__ = 'Timestamp'    
    code = Column(Integer, primary_key=True)
    timestamp = Column(BigInteger)
    date = Column(Date)
    year = Column(Integer)
    
class ClaseVino(Base):
    __tablename__ = 'ClaseVino'    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    region = Column(String)

class Wine(Base):
    __tablename__ = 'Wine'    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    tipo = Column(String)
    vintage = Column(Integer)
    bottlePrice = Column(Float)
    casePrice = Column(Float)
    claseVino = Column(Integer)

class Order(Base):
    __tablename__ = 'Order'    
    code = Column(Integer, primary_key=True)
    customer = Column(Integer)
    wine = Column(Integer)
    timeStamp = Column(Integer)
    nrBottles = Column(Integer)
    nrCases = Column(Integer)
```

```python
Base.metadata.create_all(engine)
```

# Consultas

## 1- Muestre porcentajes de tipos de vinos mas vendidos en X ao.

```python
def get_tipo_vinos_mas_vendidos_por_anio(anio):
    assert type(anio) == int, "Error"
    string_sql = """ select w.type as tipo_vino, sum(o."nrBottles") as cant_x_tipo_vino, 
                    (select sum(oo."nrBottles") from "Order" as oo join "Timestamp" as tt 
                     on oo.time = tt.code and tt.year = %d) as total_anual, 
                     (((sum(o."nrBottles"))*100)/(select sum(oo."nrBottles") 
                     from "Order" as oo join "Timestamp" as tt on oo.time = tt.code and tt.year = %d)) 
                     as porcentaje, '"+str(anio)+"' as anio from "Wine" as w join "Order" as o on (o.wine = w.code) 
                     join "Timestamp" as t on (o.time = t.code and t.year = %d) 
                     group by w.type order by cant_x_tipo_vino desc;"""
    
    return pd.read_sql(string_sql%(anio,anio, anio), engine)
    
get_tipo_vinos_mas_vendidos_por_anio(2004)
```

## 2- Cual es la temporada con mayor cantidad de ventas de X vino?


```python
def get_temporada_mayor_ventas_por_vino(code_vino):
    string_sql = """ select t.year,sum(o."nrBottles") as venta from "Wine" as w 
                    join "Order" as o on w.code = o.wine join "Timestamp" as t on t.code = o.time where w.code=10
                    group by t.year order by t.year desc;"""
    return pd.read_sql(string_sql, engine)
    
get_temporada_mayor_ventas_por_vino(10)
   
```

## 3-  Que clientes han realizado mas compras a lo largo de 4 aos?


```python
def get_buenos_clientes():
    string_sql = """ select customer as code_customer, total_ventas from (select customer, 
                SUM("nrBottles") as total_ventas from "Order" where time in (select code from "Timestamp" 
                where year > 2013) group by("customer") ) as p order by total_ventas desc;"""
    return pd.read_sql(string_sql, engine)    
    
get_buenos_clientes()
```
