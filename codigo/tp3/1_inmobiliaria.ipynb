# Definicion de tablas
![Inmobilitaria](diagramas/Inmobiliaria.png)


```python
Base = declarative_base()
class Customer(Base):
    __tablename__ = 'Customer'
    
    IDCust = Column(Integer, primary_key=True)
    Name = Column(String)
    Surname = Column(String)
    Address = Column(String)
    City = Column(String)
    Phone = Column(String)
    Budget = Column(Float)
    Office = Column(String)

class Estate(Base):
    __tablename__ = 'Estate'
    
    IDEstate = Column(Integer, primary_key=True)
    Category = Column(String)
    Area = Column(String)
    City = Column(String)
    Province = Column(String)
    Rooms = Column(Integer)
    Bedrooms = Column(Integer)
    Garage = Column(String)
    Meters = Column(Integer)

class Visit(Base):
    __tablename__ = 'Visit'
    
    IDVisit = Column(Integer, primary_key=True)
    Date = Column(Date)
    Duration = Column(Integer)

class Sale(Base):
    __tablename__ = 'Sale'
    
    IDSale = Column(Integer, primary_key=True)
    Date = Column(Date)
    AgreedPrice = Column(Float)
    Price = Column(Float)
    Status = Column(String)
    Time = Column(Integer)

class Agenda(Base):
    __tablename__ = 'Agenda'
    
    IDAgenda = Column(Integer, primary_key=True)
    Data = Column(String)
    Hour = Column(String)
    ClienteName = Column(String)

class Inmobiliaria(Base):
    __tablename__ = 'Inmobiliaria'
    
    ID = Column(Integer, primary_key=True)
    IDEstate = Column(Integer)
    IDCust = Column(Integer)
    IDVisit = Column(Integer)
    IDSale = Column(Integer)
    IDAgenda = Column(Integer)
    Date = Column(Date)
```

```python
Base.metadata.create_all(engine)
```

### 1- Â¿Que tipo de propiedad se vendio por el precio mas alto con respecto a cada ciudad y meses?

```python
def get_propiedad_precio_ciudad_mes():
   
# trelew enero tipo

    sql = """
    select City, mes, Category, precio from (
        select City, mes, Category, precio, rank() over propiedades_rank
            from( 
                    select e."City" as City, 
                    EXTRACT(MONTH FROM to_date(s."Date",'YYYY/MM/DD')) as mes,
                    EXTRACT(YEAR FROM to_date(s."Date",'YYYY/MM/DD')) as anio,
                    e."Category" as Category, 
                    MAX(s."AgreedPrice") as precio
                    from "HechoInmobiliaria" as hi 
                    inner join "Sale" as s on hi."IDSale" = s."IDSale" 
                    inner join "Estate" as e on hi."IDEstate" = e."IDEstate" 
                    group by mes, anio, e."Category",  e."City"
                    order by e."City", mes asc, precio desc
                ) as propiedades_tipo
                WINDOW propiedades_rank AS (partition by City, mes order by precio desc)
        ) as compradores_rankeados
    WHERE rank = 1 order by City, mes asc; 
    """
    
    
    resultado  = pd.read_sql(sql, engine)
    return resultado

get_propiedad_precio_ciudad_mes()
```

### 2- Â¿Quien ha comprado un piso con el precio mas alto con respecto a cada mes?

```python
def get_propiedad_precio_alto_mes():
 
    # selecciono quienes compraron un piso con el precio mas alto
    sql1 = """
    select EXTRACT(MONTH FROM to_date(s."Date",'YYYY/MM/DD')) as mes, 
                    c."Name" as Name, c."Surname" as Surname, e."Category" as Category, s."AgreedPrice" as AgreedPrice  
                    from "HechoInmobiliaria" as hi
                    inner join "Sale" as s on hi."IDSale" = s."IDSale" 
                    inner join "Estate" as e on hi."IDEstate" = e."IDEstate" 
                    inner join "Customer" as c on hi."IDCust" = c."IDCust" 
                    where e."Category" = 'piso'
                    order by s."AgreedPrice" DESC
    """
    
    # calculo con la funcion agregada rank sobre compradores_mes, que mas abajo con la clausula 
    # WINDOW  la defino sobre que se va a ejecutar, en este caso rankeo los compradores por mes
    #(los que tiene el mismo mes los pongo en un ranking por precio de mayor a menor)
    sql2 = """
    
    select mes, Name, Surname, AgreedPrice, rank() over compradores_mes
            from( 
                    select EXTRACT(MONTH FROM to_date(s."Date",'YYYY/MM/DD')) as mes, 
                    c."Name" as Name, c."Surname" as Surname, e."Category" as Category, s."AgreedPrice" as AgreedPrice  
                    from "HechoInmobiliaria" as hi
                    inner join "Sale" as s on hi."IDSale" = s."IDSale" 
                    inner join "Estate" as e on hi."IDEstate" = e."IDEstate" 
                    inner join "Customer" as c on hi."IDCust" = c."IDCust" 
                    where e."Category" = 'piso'
                    order by s."AgreedPrice" DESC
                ) as compradores WINDOW compradores_mes AS (partition by mes order by AgreedPrice DESC)
    """
    
    
    # Y finalmente solo me quedo con el numero 1 del ranking de cada mes
    sql3 = """
    select mes, concat(Name,' ', Surname) as nombre, AgreedPrice as Precio from (
        select mes, Name, Surname, AgreedPrice, rank() over particion_por_mes
            from( 
                    select EXTRACT(MONTH FROM to_date(s."Date",'YYYY/MM/DD')) as mes, 
                    c."Name" as Name, c."Surname" as Surname, e."Category" as Category, s."AgreedPrice" as AgreedPrice  
                    from "HechoInmobiliaria" as hi
                    inner join "Sale" as s on hi."IDSale" = s."IDSale" 
                    inner join "Estate" as e on hi."IDEstate" = e."IDEstate" 
                    inner join "Customer" as c on hi."IDCust" = c."IDCust" 
                    where e."Category" = 'piso'
                    order by s."AgreedPrice" DESC
                ) as compradores
                WINDOW particion_por_mes AS (partition by mes order by AgreedPrice DESC)
        ) as compradores_rankeados
    WHERE rank = 1; 
    """
    
    resultado  = pd.read_sql(sql3, engine)
    return resultado

get_propiedad_precio_alto_mes()
```

### 3- Â¿Cual es la duracion media de visitas en las propiedades de cada categorÄ±a?

```python
def get_tiempo_prom_visitas_x_cat():
   
    string_sql = """
    select e."Category", sum(v."Duration")as tiempo_total, 
    (select count(*) from "Estate" as eee where eee."Category" = e."Category" limit 1) as cant_visitas,
    (sum(v."Duration")/(select count(*) from "Estate" as ee where ee."Category" = e."Category" limit 1)) as Promedio 
    from "HechoInmobiliaria" as hi 
    join "Visit" as v on hi."IDVisit" = v."IDVisit" 
    join "Estate" as e on hi."IDEstate" = e."IDEstate" 
    group by e."Category" order by e."Category" asc;
    """
    
    resultado  = pd.read_sql(string_sql, engine)
    return resultado

get_tiempo_prom_visitas_x_cat()
```
